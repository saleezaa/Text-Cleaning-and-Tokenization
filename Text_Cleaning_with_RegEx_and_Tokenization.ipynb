# **Text Cleaning with RegEx and Tokenization**

```python
import re
import os
import pandas as pd
from collections import Counter
import kagglehub
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sunilthite/text-document-classification-dataset")

file = os.path.join(path, 'df_file.csv')
df = pd.read_csv(file)
data = df.head(1000)
```

```python
def clean_text(text):
  if not isinstance(text, str):
    return ""
  dtx = text.lower()
  dtx = re.sub(r"@[A-Za-z0-9_]+", "", dtx)
  dtx = re.sub(r"http\S+", "", dtx)
  dtx = re.sub(r"www.\S+", "", dtx)
  dtx = re.sub(r"[^a-z0-9\s]", " ", dtx)
  dtx = re.sub(r"\s+", " ", dtx).strip()
  return dtx
```

```python
data.loc[:, 'clean_text'] = data['Text'].apply(clean_text)
```
